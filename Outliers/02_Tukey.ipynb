### Método de Tukey

```python
import numpy as np

def outliers_Tukey(serie):
    q1, q3 = np.percentile(serie, [25, 75])

    IQR = q3 - q1

    lim_sup = q3 + 1.5*IQR
    lim_inf = q1 - 1.5*IQR

    outliers = serie[(serie.to_numpy() < lim_inf) | (serie.to_numpy() > lim_sup)]
    return outliers
```
