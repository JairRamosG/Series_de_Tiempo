### Técnicas de filtrado

```python
import pandas as pd

weather = pd.read_csv('datos/weather_dataset.csv',
                        parse_dates = ['datetime'],
                        index_col = 'datetime')
weather.head()
```

Suponer que se quieren los registros posteriores al 3 de Enero a las 6 pm y anteriores al 4 de Enero a las 11 am

```python
inicio = pd.to_datetime('2009-01-03 16')
final = pd.to_datetime('2009-01-04 11')

#weather.loc[inicio:final]
weather.loc[(weather.index >= inicio) & (weather.index <= final)]
```

Un intervalo de las primeras 8 horas del dia

```python
inicio = pd.to_datetime('2009-01-01 00')

delta = pd.Timedelta(hours = 8)

final = inicio + delta

weather.loc[inicio:final]
```

Cuando no interesan todas las columnas

```python
inicio = pd.to_datetime('2009-01-01 00')

delta = pd.Timedelta(hours = 8)

final = inicio + delta

#weather.loc[(weather.index >= inicio) & (weather.index <= final), ['VPmax (mbar)', 'p (mbar)', 'Tdew (degC)']]
weather.loc[inicio:final, ['VPmax (mbar)', 'p (mbar)', 'Tdew (degC)']]
```

### Ejercicio práctico

Determinar los días/horas para las cuales la temperatura sobrepasó el promedio +2 desviaciónes estándar tomando como referencia los primeros 15 días de Enero usando formato 'Monday 12, 15:00'

```python
weather = pd.read_csv('datos/weather_dataset.csv',
                        parse_dates = ['datetime'],
                        index_col = 'datetime')
weather.head()
```

```python
inicio = pd.to_datetime('2009-01-01 00')
delta = pd.Timedelta(days = 15)
final = inicio + delta

temps = weather.loc[:'2009-01-15', 'T (degC)']
temp_umbral = temps.mean() +  2*temps.std()
print(f'T umbral: {temp_umbral}')
temps[temps >= temp_umbral]
```

```python
fechas_horas = temps[temps >= temp_umbral].index
for item in fechas_horas:
    print(item.strftime('%A %d, %H:%M'))
```
