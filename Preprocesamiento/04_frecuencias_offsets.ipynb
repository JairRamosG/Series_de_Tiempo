### Frecuencia

Distancia temporal que existe entre muestras consecutivas de una serie de tiempo (Espacio entre cada muestra)



### Offset

Es la forma como Pandas representa internamente la distancia temporal entre registros

```python
# Crear un indice DatetimeIndex con 30 dias habiles a partir del 3 de septiembre de 2023

import  pandas as pd
ind1 = pd.date_range('2023-09-03', periods = 30, freq = 'B')
ind1
```

```python
ind1[10] - ind1[9]
```

```python
ind1[1] - ind1[0]
```

```python
# Crear un DatetimeIndex con periodicidad semanal empezando un miercoles

ind2 = pd.date_range('2023-09-03', periods=8, freq = 'W-WED')
ind2
```

```python
ind1[2] - ind1[0]
```

```python
# Crear un DatetimeIndex con periodicidad de una nora y 20 periodos

ind3 = pd.date_range('2023-09-03', periods=20, freq = '1h')
ind3
```

```python
# Crear un DatetimeIndex con periodicidad de una Hora y 20 minutos con 20 periodos

ind3 = pd.date_range('2023-09-03', periods=20, freq = '1h20min')
ind3
```
