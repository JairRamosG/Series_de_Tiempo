### Desplazamiento en el tiempo

Recorrer los valores en el tiempo para decidir cuando van a iniciar en otro punto diferente

```python
import numpy as np
import pandas as pd

valores = np.arange(10)
indice = pd.date_range('2023-01-01', periods=10, freq = '24h')
df = pd.DataFrame(data = {'valores': valores}, index = indice)
df
```

```python
df.shift(5) # Numero de filas que queremos desplazar
```

```python
df.shift(-3)
```

para recorrer el indice y mantener los valres, entonces se usa el parpametro de frecuencias

```python
df.shift(2, freq = '24h')
```
