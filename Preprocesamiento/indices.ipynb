## Índices

#### Crear una serie con el tipo de índice DatetimeIndex

```python
import pandas as pd

ind = pd.DatetimeIndex(['2023-09-01', '2023-09-02', '2023-09-03', '2023-09-04'])

ventas = [2350, 3150, 4200, 1175]

serie = pd.Series(ventas, index = ind)
serie
```

```python
serie.index
```

```python
serie.index[3]
```

### Crear un DataFrame

```python
ventas2 = [1923, 2413, 1097, 3213]

dicc = {'suc1' : ventas, 'suc2' : ventas2}

df_ventas = pd.DataFrame.from_dict(data = dicc).set_index(ind)
df_ventas.head()
```

```python
df_ventas.index
```

### Ejemplo práctico

```python
temps = pd.read_csv('datos/serie_temps.csv')
temps.head()
```

```python
temps['fechas'].dtypes
```

```python
# Fechas y horas separadas y con formatos object
temps = pd.read_csv('datos/serie_temps.csv',
                    parse_dates = ['fechas', 'horas'])
temps.head()
```

```python
# Combinar fechas y horas en formato adecuado
temps = pd.read_csv('datos/serie_temps.csv')

formato = '%m/%d/%Y %H:%M'
temps['fecha_hora'] = pd.to_datetime(temps['fechas'] + ' ' + temps['horas'], format = formato)
temps.head()
```

```python
temps = temps.set_index('fecha_hora')
temps = temps.drop(columns = ['fechas', 'horas'])
temps.head()
```

### Representaciónes alternativas - PeriodIndex y TimeDeltaIndex

### PeriodIndex

Expresar el indice de la serie de tiempo como intervalos de tiempo regulares (semanas, meses, años)

```python
import pandas as pd

weather = pd.read_csv('datos/weather_dataset.csv',
                        parse_dates = ['datetime'],
                        index_col = 'datetime')
weather.head()
```

```python
ind = weather.index
ind
```

```python
ind.to_period('W')
```

```python
ind.to_period('Y')
```

### TimeDeltaIndex

Representa el tiempo que ha transcurrido con respecto a un tiempo de referencia

```python
ind - ind[0]
```
