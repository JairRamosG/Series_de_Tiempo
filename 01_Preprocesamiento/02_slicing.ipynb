### Técnicas de slicing
Extraer una porción de la serie tomando como referencia la variable temporal

```python
import pandas as pd

weather = pd.read_csv('datos/weather_dataset.csv',
                        parse_dates = ['datetime'],
                        index_col = 'datetime')
weather.head()
```

```python
weather.loc['2009-01-01 02:00:00']
```

```python
weather.loc['2009-01-01 15:00:00':'2009-01-02 06:00:00']
```

```python
weather.loc['2009-01-01':'2009-01-02']
```

```python
weather.loc['2009-02']
```

```python
weather.loc['2009']
```
