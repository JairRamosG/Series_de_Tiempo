### Intentar determinar la frecuencia de la serie con infer_freq()

Sólo funciona cuando la frecuencia es regular en los registros de la serie

```python
pd.infer_freq(weather.index) # se infiere que la frecuencia es horaria
```

### asfreq()   

1.- Tomando la primera muestra de cada frecuencia

2.- Elimina datos de la serie restantes

### resample() 

1.- Agrupando periodos de tiempo indicados

2.- Después aplica una función de agregación

### Remuestreo con asfreq()

Pasar de una frecuencia de 24 por hora -> 1 por dia

```python
weather.asfreq('D') # Se ve que solo se conservan los datos del primer dia
```

```python
weather.asfreq('D').index.freq # Se indica que la frecuencia se cambio a dias
```

Primer muestra del Lunes de cada semana

```python
weather.asfreq('W-MON')
```

```python
weather.asfreq('W-MON').index.freq
```

Muestras cada 2 horas

```python
weather.asfreq('2h')
```

```python
weather.asfreq('2h').index.freq
```

Siempre se está pasando a una frecuencia de mayor tamaño, porque hay datos disponibles, pero si se quire pasar a una frecuencia menor, no van a existir esos datos y se van a inventar con espacios vacíos.

Ejemplo pasando de frecuencia de 1 hora -> media hora

```python
weather.asfreq('30min')
```

```python
weather.asfreq('30min').index.freq
```

Se pueden rellenar con valores replicados:

* Futuros: 'ffill'

* Pasados: 'bfill'

```python
weather.asfreq('30min', method='ffill')
```

```python
weather.asfreq('30min', method='bfill')
```

### Remuestreo con resample()

```python
weather.resample('24h').mean() # Toma el primer bloque de 24 hrs, lo promedia y eso es un nuevo registro 
```

### Ejercicio práctico

Tomar sólo la variable temperatura T (degC) y por cada periodo de 24 horas mostrar el promedio, el máximo y el mínimo de la serie de tiempo


```python
weather = pd.read_csv('datos/weather_dataset.csv',
                        parse_dates = ['datetime'],
                        index_col = 'datetime')
weather = weather.loc[:, ['T (degC)']]
weather.head()
```

```python
weather.resample('24h').agg(['mean', 'min', 'max'])
```

```python
def mean_min_max(serie):
    valores = [serie.mean(), serie.min(), serie.max()]
    variables = ['mean', 'min', 'max']
    return pd.DataFrame(data = valores, index = variables)

weather.resample('24h').apply(mean_min_max)
```
