# Análisis exploratorio PIB

Registro anual desde 1960 de PIB (GDP) para todos los países del mundo.
> PIB: es el monto total en USD de los bienes y servcios generados por un país en un año determinado

```python
import pandas as pd
```

```python
pib = pd.read_csv('datos/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_397725.csv',
                skiprows=4)
pib
```

```python
metadata = pd.read_csv('datos/Metadata_Country_API_NY.GDP.MKTP.CD_DS2_en_csv_v2_397725.csv')
metadata
```

### Planteamiento del análisis exploratorio

> Cómo ha sido el comportamiento del PIB a nivel mundial por regiones y por nivel de ingreso?

## Preparación de los datos

```python
pib = pib.drop(columns = ['Indicator Name', 'Indicator Code', '2023', 'Unnamed: 68'])
pib
```

```python
metadata = metadata.drop(columns=['SpecialNotes', 'TableName', 'Unnamed: 5'])
metadata
```

```python
pib = pib.merge(metadata, how='left', on='Country Code')
pib = pib.drop(columns = ['Country Name', 'Country Code'])
pib
```

```python
pib.shape
```

En este punto, el DataFrame 'pib' contiene:

* Información de 266 países
* Registro histórico de 63 años
* Nivel de ingreso de cada país

Se require de otros DataFrames:

1. Información a nivel mundial
2. Información a nivel regional
3. Información a nivel de ingreso

### Nivel mundial

```python
df_mundial = pib.iloc[:,:-2].sum(axis=0).to_frame('PIB_MUNDIAL')
df_mundial.index = pd.to_datetime(df_mundial.index, format='%Y')
df_mundial.index.name = 'Year'
df_mundial
```

```python
df_mundial.index
```

### Nivel regional

```python
pib['Region'].value_counts()
```

```python
df_region = pib.loc[:, pib.columns != 'IncomeGroup'].groupby('Region').sum()
df_region = df_region.transpose()
df_region.index = pd.to_datetime(df_region.index, format = '%Y')
df_region.index.name = 'Year'
df_region
```

### Nivel de ingreso

```python
pib['IncomeGroup'].value_counts()
```

```python
df_ningreso = pib.loc[: , pib.columns != 'Region'].groupby('IncomeGroup').sum()
df_ningreso = df_ningreso.T
df_ningreso.index = pd.to_datetime(df_ningreso.index, format='%Y')
df_ningreso.index.name = 'Time'
df_ningreso
```

### Datos faltantes

```python
print(' > Nivel Mundial')
print(f'datos: {len(df_mundial)}')
print(df_mundial.isna().sum())
print('*'*30)
print(' > Nivel Region')
print(f'datos: {len(df_region)}')
print(df_region.isna().sum())
print('*'*30)
print(' > Nivel de Ingreso')
print(f'datos: {len(df_ningreso)}')
print(df_ningreso.isna().sum())
```

## Análisis exploratorio

#### Nivel Mundial

```python
ax = df_mundial.plot(figsize=(12, 4))
ax.set_title('Nivel mundial')
ax.set_ylabel('PIB (USD)')
ax.grid(True)
```

* A nivel mundial para el PIB se tiene una serie de tiempo con tendencia ascendente
* No se presentan ciclos ni comportamientos estacionales

#### Nivel regional

```python
ax = df_region.plot(figsize=(12, 4))
ax.set_title('Nivel regional')
ax.set_ylabel('PIB (USD)')
ax.grid(True)
```

* Para todas las regiones del mundo se observa un crecimiento
* El este de Asia, Europa y Norte América presentan un mayor crecimiento

```python
totales_anuales = df_region.sum(axis = 1)
df_region_pct = 100 * df_region.div(totales_anuales, axis = 0)

ax = df_region_pct.plot(figsize=(12, 4))
ax.set_title('Nivel regional por porcentajes')
ax.set_ylabel('% PIB (USD)')
ax.set_ylim([-2, 70])
ax.legend(loc = 'best', ncols = 3)
ax.grid(True)
```

* Las regiones de Latinoamérica, Medio Este y Norte de África y Sur de Asia tienen cada una un porcentaje de aproximadamente 5%
* Norte América ha disminuido su porcentaje pasando de poco más de 50% en 1960 a casi 30% en el 2022
* El terreno perdido por Norte América ha sido ganado por Europa y el Este de Asia

#### Nivel del Ingreso

```python
ax = df_ningreso.plot(figsize=(12, 4))
ax.set_title('Nivel de Ingreso')
ax.set_ylabel('PIB (USD)')
ax.grid(True)
```

* A partir del año 1970 se tuvo una importante diferencia de crecimiento entre los niveles de ingreso altos con el resto
* Para el año 2022, el nivel de ingreso medio es poco menos de la mitad que el nivel de ingreso alto
* El nivel de ingreso bajo se ha mantenido constante desde 1960

```python
df_ningreso
```

```python
totales_anuales = df_ningreso.sum(axis = 1)
df_ningreso_pcte = 100 * df_ningreso.div(totales_anuales, axis = 0)

ax = df_ningreso_pcte.plot(figsize=(12, 4))
ax.set_title('Nivel de Ingreso por porcentajes')
ax.set_ylabel('% PIB (USD)')
ax.grid(True)
```

* A partir del año 200, el porcentaje del nivel de ingreso alto ha bajado hasta un 60%
* Desde el año 1990 el porcentaje del nivel de ingreso  medio ha tenido una tendencia positiva
* El porcentaje del ingreso bajo siempre ha sido constate desde 1960

### Tasas de crecimiento a nivel mundial

```python
tasa_mundial = 100 * df_mundial.diff(1).div(df_mundial, axis=0)

ax = tasa_mundial.plot(figsize=(12, 4))
ax.set_title('Tasa de crecimiento a nivel mundial')
ax.set_ylabel('% PIB (USD)')
ax.grid(True)
```
