# Análisis exploratorio PIB

Registro anual desde 1960 de PIB (GDP) para todos los países del mundo.
> PIB: es el monto total en USD de los bienes y servcios generados por un país en un año determinado

```python
import pandas as pd
```

```python
pib = pd.read_csv('datos/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_397725.csv',
                skiprows=4)
pib
```

```python
metadata = pd.read_csv('datos/Metadata_Country_API_NY.GDP.MKTP.CD_DS2_en_csv_v2_397725.csv')
metadata
```

### Planteamiento del análisis exploratorio

> Cómo ha sido el comportamiento del PIB a nivel mundial por regiones y por nivel de ingreso?

### Preparación de los datos

```python
pib = pib.drop(columns = ['Indicator Name', 'Indicator Code', '2023', 'Unnamed: 68'])
pib
```

```python
metadata = metadata.drop(columns=['SpecialNotes', 'TableName', 'Unnamed: 5'])
metadata
```

```python
pib = pib.merge(metadata, how='left', on='Country Code')
pib = pib.drop(columns = ['Country Name', 'Country Code'])
pib
```

```python
pib.shape
```

En este punto, el DataFrame 'pib' contiene:

* Información de 266 países
* Registro histórico de 63 años
* Nivel de ingreso de cada país

Se require de otros DataFrames:

1. Información a nivel mundial
2. Información a nivel regional
3. Información a nivel de ingreso

### Nivel mundial

```python
df_mundial = pib.iloc[:,:-2].sum(axis=0).to_frame('PIB_MUNDIAL')
df_mundial.index = pd.to_datetime(df_mundial.index, format='%Y')
df_mundial.index.name = 'Year'
df_mundial
```

```python
df_mundial.index
```

### Nivel regional

```python
pib['Region'].value_counts()
```

```python
df_region = pib.loc[:, pib.columns != 'IncomeGroup'].groupby('Region').sum()
df_region = df_region.transpose()
df_region.index = pd.to_datetime(df_region.index, format = '%Y')
df_region.index.name = 'Year'
df_region
```

### Nivel de ingreso

```python
pib['IncomeGroup'].value_counts()
```

```python
df_ningreso = pib.loc[: , pib.columns != 'Region'].groupby('IncomeGroup').sum()
df_ningreso = df_ningreso.T
df_ningreso.index = pd.to_datetime(df_ningreso.index, format='%Y')
df_ningreso.index.name = 'Time'
df_ningreso
```

### Datos faltantes

```python
print(' > Nivel Mundial')
print(f'datos: {len(df_mundial)}')
print(df_mundial.isna().sum())
print('*'*30)
print(' > Nivel Region')
print(f'datos: {len(df_region)}')
print(df_region.isna().sum())
print('*'*30)
print(' > Nivel de Ingreso')
print(f'datos: {len(df_ningreso)}')
print(df_ningreso.isna().sum())
```
