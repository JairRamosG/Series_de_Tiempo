### Imputación con interpolacion

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def graficar_serie(df, obs = '', titulo=''):
    '''
    Graficar la serie y ver sus valores extremos
    Params:
        - df    : DataFrame de Pandas con la serie de tiempo
        - obs   : Nombre de la columna a graficar
        - titulo: Título del gráfico
    '''

    fig, ax = plt.subplots(1, 1)

    if type(df) == pd.DataFrame:
        ax = sns.pointplot(data = df, x = df.index, y = obs, markers='')
    else:
        df = df.to_frame(name = obs)
        ax = sns.pointplot(data = df, x = df.index, y = obs, markers='')
    
    n_ticks = 10
    locator = mdates.AutoDateLocator(maxticks = n_ticks)
    ax.xaxis.set_major_locator(locator)

    plt.xticks(rotation = 45)
    
    ax.set_xlabel('Tiempo')
    ax.set_ylabel(obs)
    ax.set_title(titulo) 
```

### Imputación ***interpolate()***



* linear
* quadratic
* cubic
* nearest
* spline (order)

```python
co2 = pd.read_csv('datos/co2_faltantes.csv',
                        parse_dates=['año'],
                        index_col='año')
```

```python
metodos = ['linear', 
            'quadratic',
            'cubic', 
            'nearest',
            'spline']

for metodo in metodos:
    if metodo != 'spline':
        co2[metodo] = co2['co2'].interpolate(method=metodo)
    else:
        co2[metodo] = co2['co2'].interpolate(method=metodo, order = 3)
co2
```

```python
graficar_serie(co2, obs = 'co2', titulo = 'co2')
graficar_serie(co2, obs = 'linear', titulo = 'Interpolación linear')
graficar_serie(co2, obs = 'quadratic', titulo = 'Interpolación quadratic')
graficar_serie(co2, obs = 'cubic', titulo = 'Interpolación cubic')
graficar_serie(co2, obs = 'nearest', titulo = 'Interpolación nearest')
graficar_serie(co2, obs = 'spline', titulo = 'Interpolación spline')
```
